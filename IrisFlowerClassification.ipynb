```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
print(iris_df.head())
```

```python
print("missing values:\n", iris_df.isnull().sum())
```

```python
encode = LabelEncoder()
iris_df['target'] = encode.fit_transform(iris_df['target'])
print("encoded dataframe: /n",iris_df.head())
```

```python
scaler = StandardScaler()
iris_df[iris.feature_names] = scaler.fit_transform(iris_df[iris.feature_names])
print("scaled dataframe:\n", iris_df.head())
```

```python
x = iris_df[iris.feature_names]
y = iris_df['target']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print("tanning set size: ", x_train.shape)
print("testing set size: ", x_test.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
for i , feature in enumerate(iris.feature_names):
    plt.subplot(2,2,i+1)
    sns.histplot(data=iris_df, x=feature, hue='target', kde=True)
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data=iris_df, hue='target', markers=['o','s','D'])
plt.show()
```
